```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import time
pd.set_option('display.max_columns', None)
```

```python
# Load Dataset
df = pd.read_csv('dataset_clean.csv')
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df
```

```python
# Check how many nans are in each column
df.isna().sum()
```

```python
# find row of df where 'SP Name' is Kyle Lohse
df[df['SP Name'] == 'Kyle Lohse']
```

```python
# Loop through rows of df where SO is nan and P_mlb_id is not nan
from pybaseball import pitching_stats_range
for i in df[(df['SP SO'].isna()) & (df['P_mlb_id'].notna())].index:
    row = df.loc[i]
    pitcher_mlb_id = str(int(row['P_mlb_id']))
    date = str(row['Date'])
    date = date[0:4] + '-' + date[4:6] + '-' + date[6:8]
    time.sleep(1)
    print('getting new stats ' + date + ' ' + 'index: ' + str(i))
    pitcher_stats = pitching_stats_range(date,)
    pitcher_game_stats = pitcher_stats.loc[pitcher_stats['mlbID'] == pitcher_mlb_id]
    if not pitcher_game_stats['SO'].empty:
        pitcher_strikeouts = pitcher_game_stats['SO'].values[0]
        #convert pitcher_strikeouts to numpy float64
        pitcher_strikeouts = np.float64(pitcher_strikeouts)
        print(type(pitcher_strikeouts))
        print(pitcher_game_stats['Name'].values[0] + ' ' + str(pitcher_strikeouts))
        df.loc[i, 'SP SO'] = pitcher_strikeouts
    if i % 5 == 0:
        df.to_csv('game_logs_raw.csv')
```
