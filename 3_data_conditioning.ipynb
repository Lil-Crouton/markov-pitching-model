```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
game_logs_df = pd.read_csv('2_game_logs_raw.csv', index_col=0)
```

```python
# Drop columns that are not needed
gl_df = game_logs_df.drop(columns=['mlb_id','Date','Visiting Team','Visiting Team G#',\
                                'Home Team','Home Team G#','SP ID','SP Name',\
                                'P1 ID','P2 ID','P3 ID','P4 ID','P5 ID','P6 ID',\
                                'P7 ID','P8 ID','P9 ID','P_mlb_id','P_wKN/C','P_KNv',\
                                'P_KN%','wKN/C','Away Score','Home Score'])
gl_df
```

```python
# Drop rows that are not needed
gl_df = gl_df[gl_df['P_FBv'].notna()]
gl_df = gl_df[gl_df['P_wFB/C'].notna()]
```

```python
# Create trimmed dataframe
gl_df_trim = gl_df.drop(columns=['P_SL%','P_SLv','P_CT%','P_CTv','P_CB%','P_CBv',\
                                 'P_CH%','P_CHv','P_SF%','P_SFv','P_wSL/C','P_wCT/C',\
                                 'P_wCB/C','P_wCH/C','P_wSF/C'])
```

```python
# Add columns for pitch types
gl_df['P_Throws_CB'] = ~gl_df['P_CBv'].isna()
gl_df['P_Throws_SL'] = ~gl_df['P_SLv'].isna()
gl_df['P_Throws_CH'] = ~gl_df['P_CHv'].isna()
gl_df['P_Throws_CT'] = ~gl_df['P_CTv'].isna()
gl_df['P_Throws_SF'] = ~gl_df['P_SFv'].isna()


gl_df = gl_df.replace(False, 0)
gl_df = gl_df.replace(True, 1)
```

```python
# Fill in missing values with median
gl_df['P_SL%'].fillna(gl_df['P_SL%'].median(), inplace=True)
gl_df['P_CT%'].fillna(gl_df['P_CT%'].median(), inplace=True)
gl_df['P_CB%'].fillna(gl_df['P_CB%'].median(), inplace=True)
gl_df['P_CH%'].fillna(gl_df['P_CH%'].median(), inplace=True)
gl_df['P_SF%'].fillna(gl_df['P_SF%'].median(), inplace=True)

gl_df['P_SLv'].fillna(gl_df['P_SLv'].median(), inplace=True)
gl_df['P_CTv'].fillna(gl_df['P_CTv'].median(), inplace=True)
gl_df['P_CBv'].fillna(gl_df['P_CBv'].median(), inplace=True)
gl_df['P_CHv'].fillna(gl_df['P_CHv'].median(), inplace=True)
gl_df['P_SFv'].fillna(gl_df['P_SFv'].median(), inplace=True)

gl_df['P_wSL/C'].fillna(gl_df['P_wSL/C'].median(), inplace=True)
gl_df['P_wCT/C'].fillna(gl_df['P_wCT/C'].median(), inplace=True)
gl_df['P_wCB/C'].fillna(gl_df['P_wCB/C'].median(), inplace=True)
gl_df['P_wCH/C'].fillna(gl_df['P_wCH/C'].median(), inplace=True)
gl_df['P_wSF/C'].fillna(gl_df['P_wSF/C'].median(), inplace=True)
```
